1.Make a small generator program function returning armstrong numbers in between 1 to 1000.

```
lst=list(range(1, 1000+1))
```

```
 
 
def armstrong_number(lst):
    for item in lst:
       sum=0
       num=item
       while num>0:
           d=num%10
           sum+=d**3
           num//=10
       if(item==sum):
           yield item
```

```
list(armstrong_number(lst))
```

```
 %%writefile prime.py
 
'''
Creating a python file for finding whethere the given number is prime number or not.
'''
 
def is_prime(given_no):
     if given_no >= 1:
         for i in range(2, given_no):
             if (given_no % i) == 0:
                 final_result="Not Prime"
             else:
                 final_result="Prime"
     else:
         final_result="Not Prime"
     return final_result
```

```
!pylint "prime.py"
```

```
%%writefile primeUnitTest.py
 
import unittest
import prime
 
class testPrime(unittest.TestCase):
    def testSinglePrime(self):
        a = 3
        result = prime.is_prime(a)
        self.assertEquals(result, "Prime")
        
    def testingDoublePrime(self):
        b = 11 
        result = prime.is_prime(b)
        self.assertEquals(result, "Prime")
 
if __name__ == "__main__":
    unittest.main()
```
