```
 def get_input(range_fun):
    def wrap_function():
       print("Here are your prime numbers")
       a=int(input("Enter lower number: "))
       b=int(input("Enter higher number: "))
       range_fun(a, b)
    return wrap_function
```

```
 
@get_input
def prime_numbers(low, high):
    for i in range(low, high+1):
        for j in range(2, i):
            if(i%j==0):
               break
        else:
           print(i)
```
