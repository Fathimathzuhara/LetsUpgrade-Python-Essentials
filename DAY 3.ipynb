 1.You all are pilots, you have to land a plane,the altitude required for landing a plane is 1000ft,if it is less than that tell pilot to land the plane, if it is more than1000ft and less than 5000ft tell "Come down to 1000ft", if it is more than 5000ft, tell"go around and try later"

```
altitude =int(input("Enter altitude in ft: "))
```

```
if altitude <=1000:
    print("Safe land")
elif altitude >1000 and altitude <=5000:
    print("come down to 1000ft")
else:
    print("Go around and try later!!")
```
