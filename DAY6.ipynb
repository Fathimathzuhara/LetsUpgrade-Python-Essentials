 1.Create a bank account that has two attributes,owner name, balance and two methods deposit and withdraw

```
 
 
class bankaccount:
   def __init__(self, ownername, balance):
       self.ownername=ownername
       self.balance=balance
   def deposit(self):
      deposit=int(input("Enter amount to be deposited: "))
      self.balance=self.balance+deposit
      print("Amount deposited is ", deposit)
   def withdraw(self):
       withdraw=int(input("Enter amount to  be withdrawn:"))
       if (withdraw>self.balance):
         print("please enter amount less than your balance")
       else:
         self.balance=self.balance-withdraw
         print("you have withdrew: ", withdraw)
AC=bankaccount("Raj", 5000)
print("Account holder :", AC.ownername)
print("Initial account balance is: ", AC.balance)
AC.deposit()
AC.withdraw()
print("Net available balane is: ",AC.balance)
```

 2.Create a class of cone that has two attributes R=radius,h=height and two methods 
Volume=pi×r^2=(h/3)
Surface area:base:pi×r^2, side:pi×r×(r^2+h^2)^1÷2

```
 import math
 
pi = math.pi
 
class cone:
    
    def __init__(self,r,h): 
 
        self.r=r
        
        self.h=h
        
    def volume(self):
        
        result = (1 / 3) * pi * self.r * self.r * self.h
        
        print("\nVolume Of Cone is :",result)
        
    def surfacearea(self):
        
        result = pi * self.r * self.h + pi * self.r * self.r
         
        print("\nSurface Area Of Cone is :",result)
```

```
r1=float(input("Enter radius of the cone "))
h1=float(input("Enter height of the cone "))
C=cone(r1, h1)
C.volume()
C.surfacearea()
```
